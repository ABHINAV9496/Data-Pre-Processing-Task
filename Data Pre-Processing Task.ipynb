# Data Cleaning

```python
#Number of missing values in each feature
data.isnull().sum()
```

```python
#Duplicate value checking
data.duplicated().sum()
```

```python
#Replacing Gender to cateogories as male and female
data['Gender'].replace({0: 'male', 1: 'female'}, inplace=True)
data
```

```python
data['Age'].replace(to_replace=0,value=np.nan,inplace=True)
data
```

```python
data.Age.max()
```

```python
data.Age.min()
```

```python
avg_salary=data.Salary.mean()
avg_salary
```

```python
data.Salary.fillna(avg_salary, inplace=True)
data
# using fillna() function filling all null values with mean of the salary
```

```python
#removing duplicates
data.drop_duplicates()
```

```python
#finding outliers
Q1 = data.Salary.quantile(0.25)
Q3 = data.Salary.quantile(0.75)
IQR = Q3 - Q1
Q1,Q3,IQR
```

```python
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
lower_bound,upper_bound
```

# Data Analysis

```python
#Filter the data with age >40 and salary<5000
filtered_data = data[(data["Age"]>40)&(data["Salary"]<5000)]
filtered_data
```

```python
#Scatter plot with Age vs Salary
plt.figure(figsize=(10,6))
sns.scatterplot(x="Age",y="Salary",data=filtered_data)
plt.title("Age vs Salary")
plt.xlabel("Age")
plt.ylabel("Salary")
plt.show()
```

```python
# Count the number of people from each place and represent it visually?
data.Place.value_counts()
```

```python
plt.title("Number Of People From Each Place")
plt.xlabel("Place")
plt.ylabel("Count")
data.Place.value_counts().plot(kind = 'bar',color='r')
plt.show()
```
